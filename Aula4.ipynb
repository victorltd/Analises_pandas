```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#Criando dataframe
df = pd.read_excel('AdventureWorks.xlsx')
df.tail()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#Receita total
df['Valor Venda'].sum()
```

```python
df['custo'] = df['Custo Unitário'].mul(df['Quantidade'])
df.head()
```

```python
# Custo total
df.custo.sum()
```

```python
# Lucro -> receia - lucro
df['lucro']= df['Valor Venda'] - df['custo']
df.head()
```

```python
#ver a qtd de dias que leva ate enviar
df['tempo_envio']= (df['Data Envio'] - df['Data Venda']).dt.days
df.sample(5)
```

```python
#media do tempo de envio
df.groupby('Marca')['tempo_envio'].mean()
```

## Missing values


```python
#ver se tem valores ausentes
df.isnull().sum()
```

### Sabendo lucro por marcas

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#agrupar por ano e marca
df.groupby([df['Data Venda'].dt.year,'Marca'])['lucro'].sum()
```

```python
#Resentando o index 
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum().reset_index()
lucro_ano.head()
```

```python
# total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# grafico total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total produtos vendidos')
plt.xlabel('Qtde')
plt.ylabel('Produto');
```

```python
#lucro por ano
df.groupby(df['Data Venda'].dt.year)['lucro'].sum().sort_values(ascending=False).plot.bar(title= 'Lucro por ano')
plt.xlabel('Ano')
plt.ylabel('Lucro');
```

```python
#slecionando apenas os de 2009
filtro = df['Data Venda'].dt.year == 2009
df_2009= df.loc[filtro]
filtro.head()
```

```python
#plotando esse grafico
df_2009.groupby(df['Data Venda'].dt.month)['lucro'].sum().plot(title='Lucros por mes')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
#plotando lucro por marca
df_2009.groupby(df['Marca'])['lucro'].sum().plot.bar(title='Lucros por marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
